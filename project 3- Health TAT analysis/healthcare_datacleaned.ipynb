```python
df.columns = df.columns.str.strip()
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(" ", "_")
```

```python
from sqlalchemy import create_engine

engine = create_engine("mysql+pymysql://root:123456789@localhost/healthcare_tat_project")
```

```python
df.to_sql(
    name="cleaned_tat_data",
    con=engine,
    if_exists="replace",   # use 'append' if table already exists
    index=False
)
```

```python
df.select_dtypes(include=['timedelta64[ns]']).columns
```

```python
timedelta_cols = df.select_dtypes(include=['timedelta64[ns]']).columns

for col in timedelta_cols:
    df[col] = df[col].dt.total_seconds() / 3600
```

```python
df.to_sql(
    name="cleaned_tat_data",
    con=engine,
    if_exists="replace",
    index=False
)
```
